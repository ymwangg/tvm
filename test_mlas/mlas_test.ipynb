```python
from tvm import te
import tvm

def packb(K, N, B):
    f = tvm._ffi.get_global_func("tvm.contrib.mlas.gemm_packb_size")
    packb_size = f(N, K)
    arr_size = packb_size / 4
    return te.extern(
        (int(arr_size),),
        [B],
        lambda ins, outs: tvm.tir.call_packed(
            "tvm.contrib.mlas.gemm_packb",
            N, K, N,
            ins[0],
            outs[0],
        ),
        dtype='float32'
    )
```

```python
B = te.placeholder([100, 100], dtype='float32')
```

```python
packed = packb(100, 100, B)
```

```python
s = te.create_schedule([packed.op])
```

```python
mod = tvm.build(s, [B, packed])
```

```python
mod(a, b)
```

```python
def run_opt_pass(expr, opt_pass):
    assert isinstance(opt_pass, tvm.transform.Pass)

    mod = tvm.IRModule.from_expr(expr)
    mod = relay.transform.InferType()(mod)
    mod = opt_pass(mod)
    entry = mod["main"]
    return entry if isinstance(expr, relay.Function) else entry.body
```
